```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import networkx as nx


import numpy as np
from numpy.random import randint as r

df = pd.read_csv('data/London.csv')


# Make a set of all the junctions; coming from the start junctions and end junctions.
junctions = set(df.EndJunction.values)
junctions = set(df.StartJunction.values)



G = nx.Graph()


# G.add_nodes_from(df['road_id'])
G.add_nodes_from(junctions)
G.add_edges_from(df[['StartJunction','EndJunction']].values, pos=list(df[['Easting', 'Northing']].values))


nx.draw_shell(G)
```

```python
rds = list(junctions)[r(0, 30)::120]
print(rds)

road_connections_list = [
    rds[0:2],
    rds[2:4],
    rds[4:6],
    rds[6:],
]

for i in range(10):
    tmp = rds[r(3)::r(1,len(rds)) ]
    print(tmp)

road_connections = dict()
for i, road in enumerate(rds):
    if i < len(road_connections):
        road_connections[road] = road_connections_list[i]
    else:
        road_connections[road] = rds[r(3)::r(1,len(rds)) ]

    


# juncs = [road_name for road_name, road_juncts in dict_of_roads]
# egs = [np.random.random_integers(0,9) for i in juncs]
G = nx.Graph()
G.add_nodes_from(rds)
for road, connections in road_connections:
    G.add_edges_from([road,connections])
nx.draw_circular(G)
```
